### **데이터 정제**

```python
# 데이터 분할
# 독립변수 : Tiredness, Dry-Cough, Difficulty-in-Breathing, Sore-Throat,
#          None_Sympton, Pains, Nasal-Congestion, Runny-Nose, None_Experiencing
# 종속변수 : Severity_Mild
X = df.iloc[:, 0:9]
y = df['Severity_Mild']

# 학습 및 테스트 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.3, stratify = y)
```

```python
df.iloc[:,0:9]
```

```python
# random_state 전
# print('학습 데이터 수: {0}, 테스트 데이터 수: {1}'.format(len(X_train), len(X_test)))
# 학습 데이터 수: 237600, 테스트 데이터 수: 79200

# random_state 후
print('학습 데이터 수: {0}, 테스트 데이터 수: {1}'.format(len(X_train), len(X_test)))
# 학습 데이터 수: 221760, 테스트 데이터 수: 95040
# stratify하면 random_state 안 넣어도 될 듯?
```

### **의사결정 나무**

```python
# DecisionTreeClassifier 객체 생성
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state = 1)

# 학습 데이터 학습 수행
model.fit(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz
from subprocess import call
from IPython.display import Image

# 의사결정나무 모델을 dot 파일로 추출
export_graphviz(model, out_file = 'asthma_tree.dot')

# dot 파일을 .png로 변환
call(['dot', '-Tpng', 'asthma_tree.dot', '-o', 'asthma_tree.png', '-Gdpi = 600'])

# png 출력
Image(filename = 'asthma_tree.png')
```

```python
# 테스트용 데이터로 모델 평가
score = model.score(X_test, y_test)
print('테스트 정확도 {0}'.format(score))
```

### **KFold**

```python
# KFold 객체 생성
from sklearn.model_selection import KFold

kf = KFold(n_splits = 3)
for train_index, test_index in kf.split(X):
  print('-' * 50)
  print('학습용:', train_index)
  print('\n학습용 레이블:', y[train_index].unique())
  print('\n테스트용:', test_index)
  print('\n테스트용 레이블:', y[test_index].unique())
```

```python
# StartifiedKFold
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits = 3)
for train_index, test_index in skf.split(X, y):
  print('-' * 50)
  print('학습용:', train_index)
  print('\n학습용 레이블:', y[train_index].unique())
  print('\n테스트용:', test_index)
  print('\n테스트용 레이블:', y[test_index].unique())
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(model, X_train, y_train, cv = 3, return_estimator = True)
scores
```

```python
for i in range(3):
  score = scores['estimator'][i].score(X_test, y_test)
  print('{0}번째 의사결정나무 정확도: {1:.2f}'.format(i + 1, score))
```
