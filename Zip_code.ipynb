### **지역**

```python
# zip_code 값을 기반으로 지역을 변환하는 함수
def convert_zip_to_country(zip_code):
    prefix = int(zip_code.split('-')[0])  # 하이푼(-)을 기준으로 앞 3자리를 추출하고 정수로 변환
    if 100 <= prefix < 160:
        return '서울'
    elif 200 <= prefix < 270:
        return '강원'
    elif 300 <= prefix < 310:
        return '대전'
    elif prefix == 339:
        return '세종'
    elif 310 <= prefix < 358:
        return '충청남도'
    elif 360 <= prefix < 396:
        return '충청북도'
    elif 400 <= prefix < 410:
        return '인천'
    elif 410 <= prefix < 488:
        return '경기도'
    elif 500 <= prefix < 510:
        return '광주'
    elif 510 <= prefix < 557:
        return '전라남도'
    elif 560 <= prefix < 598:
        return '전라북도'
    elif 600 <= prefix < 620:
        return '부산'
    elif 620 <= prefix < 679:
        return '경상남도'
    elif 680 <= prefix < 690:
        return '울산'
    elif 690 <= prefix < 700:
        return '제주'
    elif 700 <= prefix < 710:
        return '대구'
    elif 710 <= prefix < 800:
        return '경상북도'
    else:
        return '알 수 없음'  # 범위에 포함되지 않는 경우

# 'zip_code' 열을 기반으로 'country' 열을 생성
df_zip['country'] = df_zip['zip_code'].apply(convert_zip_to_country)

# 결과 확인
df_zip
```

```python
# 각 지역 좌표 변수열 생성
# 'country' 값에 따라 'LatLon' 변수 생성
def get_lat_lon(country):
    if country == '서울':
        return '37.5665, 126.9779'  # 서울의 위도와 경도

    elif country == '강원도':
        return '37.8228, 128.1555'  # 강원도의 위도와 경도

    elif country == '대전':
        return '36.3504, 127.3845'  # 대전의 위도와 경도

    elif country == '충청남도':
        return '36.5184, 126.8000'  # 충청남도의 위도와 경도

    elif country == '세종':
        return '36.5040, 127.2494'  # 세종의 위도와 경도

    elif country == '충청북도':
        return '36.4918, 127.3925'  # 충청북도의 위도와 경도

    elif country == '인천':
        return '37.4562, 126.7052'  # 인천의 위도와 경도

    elif country == '경기도':
        return '37.4138, 127.5183'  # 경기도의 위도와 경도

    elif country == '광주':
        return '35.1595, 126.8526'  # 광주의 위도와 경도

    elif country == '전라남도':
        return '34.8679, 126.9910'  # 전라남도의 위도와 경도

    elif country == '전라북도':
        return '35.7175, 127.1530'  # 전라북도의 위도와 경도

    elif country == '부산':
        return '35.1795, 129.0756'  # 부산의 위도와 경도

    elif country == '경상남도':
        return '35.4606, 128.2132'  # 경상남도의 위도와 경도

    elif country == '울산':
        return '35.5383, 129.3113'  # 울산의 위도와 경도

    elif country == '제주':
        return '33.4890, 126.4983'  # 제주의 위도와 경도

    elif country == '대구':
        return '35.8714, 128.6014'  # 대구의 위도와 경도

    elif country == '경상북도':
        return '36.4919, 128.8889'  # 경상북도의 위도와 경도
    # 다른 지역에 대한 경우도 추가할 수 있음
    else:
        return None  # 알 수 없는 경우

# 'LatLon' 변수열 생성
df_zip['LatLon'] = df_zip['country'].apply(get_lat_lon)

# 결과 확인
df_zip
```

```python
# LatLon 변수열을 분할하여 Latitude 및 Longitude 변수열 생성
df_zip[['Latitude', 'Longitude']] = df_zip['LatLon'].str.split(', ', expand=True)

# 결과 확인
df_zip[['Latitude', 'Longitude']] = df_zip[['Latitude', 'Longitude']].astype(float)
df_zip = df_zip.drop(columns=['LatLon'])
df_zip
```

```python
# 'country' 열의 고유한 값과 해당 값의 등장 횟수를 계산
value_counts = df_zip['country'].value_counts().reset_index()

# 결과를 출력
value_counts
```
