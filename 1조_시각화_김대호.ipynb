```python
# 히트맵, 파이차트, 시계열, 바이올린

import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/computer_games_utf8.csv')
df
```

```python
df['Developer'].describe()
```

```python
# 가장 많은 개발사 상위 10개

from collections import Counter

# 변수의 빈도수 계산
variable_counts = Counter(df['Developer'])

# 빈도수가 가장 높은 상위 10개 변수 추출
tpDvlp = variable_counts.most_common(10)

# 결과 출력
for variable, count in tpDvlp:
    print(f'{variable}: 빈도수 {count}번')
```

### **개발사 파이차트**

```python
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns

# 개발사별 빈도수 계산
distribution_counts = Counter(df['Developer'])

# 빈도수가 가장 높은 상위 10개 개발사 추출
top_10_distribution = distribution_counts.most_common(10)

# 추출한 상위 10개 개발사와 그 점유율을 새로운 리스트로 저장
tp_labels = [item[0] for item in top_10_distribution]
tp_sizes = [item[1] for item in top_10_distribution]

# 파이 차트 내부 원의 반지름
inner_radius = 0.3

# 파이 조각 강조
explode = [0.05, 0, 0, 0, 0, 0, 0, 0, 0, 0]

# 라벨의 개수 만큼 색상 리스트 생성
colors = sns.color_palette('hls',len(tp_labels))

# 파이차트 그리기
plt.figure(figsize=(8, 9))
plt.pie(tp_sizes, labels = tp_labels, colors = colors, autopct='%1.1f%%', startangle=140, explode = explode,
        wedgeprops=dict(width=inner_radius, edgecolor='w'))
plt.title('Top10 Developer')
plt.axis('equal')

# 차트 표시
plt.show()
```
