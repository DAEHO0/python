```python
# 채식주의자들을 위한 묶음 상품 구성
# 육류를 소비하면서 채소를 같이 소비하는? 사람들을 위해
# 채소 품목을 선택한 이유에서 많이 언급된 단어를 추출해 이유를 밝힘?

# 데이터 불러오기
import pandas as pd
df = pd.read_csv('data2.csv', encoding = 'utf-8')
df
```

```python
# Null 값, US$ 제거를 위한 전처리

# 전처리
import pandas as pd

# NULL 값을 삭제한 데이터프레임 생성
df_cln = df.dropna()

print("\nDataFrame after dropping NULL values:")
print(df_cln)

# 'US$'를 제거 후 float 타입으로 변경
df_cln['item_price'] = df_cln['item_price'].str.replace('US\$', '', regex=True).astype(float)

print("Original DataFrame:")
print(df_cln)
```

```python
# item_name 품목 횟수 출력
grpd = df_cln.groupby('item_name').size()

# 결과
print(grpd)

# 가장 많이 언급된 품목은 'chicken bowl'
# 육류 품목이 압도적으로 많음
# 육류를 소비하는 소비자들은 채소도 같이 소비하지 않을까?
# veggie로 시작하는 품목에 집중
```

```python
# item_name 중 품목이 Veggie Burrito인 값만 출력
vbr = df_cln[df_cln['item_name'] == 'Veggie Burrito']
print(vbr)
```

```python
import pandas as pd

# Veggie Burrito 가격별 그룹화
vbr2 = vbr.groupby('item_price')

# 기술통계
vbr2.describe()

# 8.49$ 가격대가 가장 많이 팔림
```

```python
# Veggie Burrito를 산 고객들 중 해당 품목을 선택한 이유에서
# 가장 많이 언급된 단어를 추출

import pandas as pd
from collections import Counter

# 단어 추출 및 빈도수 계산
word_list = ' '.join(vbr['choice_description']).split()
word_counts = Counter(word_list)

# 빈도수가 높은 순으로 정렬
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# 상위 빈도수 단어 출력
top_words = 10
for word, freq in sorted_word_counts[:top_words]:
    print(f'{word}: {freq}')
    

# 가장 많이 언급된 단어 : 콩(Beans), 쌀(Ricd)
```

```python
# 결론 :
# 채식 상품 중 Veggie Burrito 상품이 가장 많이 팔림
# 그 중에서도 8.49$ 가격대가 많이 팔림

# Veggie Burrito 상품을 선택한(구매한) 고객들의
# 선택 이유를 가장 많이 언급된 단어를 추출함
# 결과는 Beans(콩), Rice(쌀)가 가장 많이 언급됨
# 고객이Veggie Burrito를 선택한 이유는 콩과 쌀 재료가 들어간 제품을
# 많이 선택한다고 추론해볼 수 있음

# Insight :
# 8.49$ 이하로 콩과 쌀 재료가 들어간 상품을 묶음 판매
```
