### **도서 데이터 준비**

```python
# 과연 할 수 있을까...
# 데이터 - 강원도 평창군 도서관 도서 정보(공공데이터 포털)
import pandas as pd

# 도서 데이터
# 171,081 행 X 11 열
bks = pd.read_csv('/content/drive/MyDrive/pyeongchang_library_20220915.csv', encoding = 'cp949')
bks
```

### **데이터 전처리**

```python
# NaN 값 확인
na = bks.isna().sum()

# null 값 확인
nll = bks.isnull().sum()

# NaN 값 출력
print('na:\n', na)
print('-' * 25)

# null 값 출력
print('null:\n', nll)
print('-' * 25)

# 도서명에 중복된 값이 있는지 확인
dpc = bks[bks['도서명'].duplicated()]
dpc # 없는 것 같은데...
```

```python
# NaN 값 == null 값, 중복값 29개
# 도서명은 필수로 있어야 하므로 누락된 도서명 행은 삭제

# NaN, null인 행 추출
na_rws = bks[bks.isna().any(1)]
nll_rws = bks[bks.isnull().any(axis=1)]

# 총 355 행
na_rws
```

```python
# 총 355 행
nll_rws
```

```python
# 누락된 값 삭제
bks_na = bks.dropna()

# 중복값 삭제
bks_cln = bks_na.drop_duplicates()

# 데이터 확인
na_cln = bks_cln.isna().sum()
nll_cln = bks_cln.isnull().sum()
dpc_cln = bks_cln.duplicated().sum()

print(na_cln)
print('-' * 25)
print(nll_cln)
print('-' * 25)
print('중복행 수:', dpc_cln)
```
